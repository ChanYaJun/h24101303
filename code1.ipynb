```python
#使用模型
import pandas as pd
import numpy as np
df=pd.read_csv("train.csv")
filt=df["Exited"]==1
df

```

```python
#把測試資料和驗證做標準化
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
#train的長度
trn=df_train.shape[0]
df_train = df_train.drop(['Exited'] , axis=1)
cbdf = pd.concat([df_train,df_test])
cbdf.head()
#print(cbdf.shape)

Geography=pd.get_dummies(cbdf["Geography"],prefix="Geography")
cbdf=cbdf.drop("Geography",1)
cbdf=pd.concat([Geography,cbdf],axis=1)

Gender=pd.get_dummies(cbdf["Gender"],prefix="Gender")
cbdf=cbdf.drop("Gender",1)
cbdf=pd.concat([Gender,cbdf],axis=1)

list(cbdf.columns)
cbdf=cbdf.drop("Surname",axis=1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_standard = scaler.fit_transform(cbdf[[
 'CreditScore',
 'Age',
 'Tenure',
 'Balance',
 'NumOfProducts',
 'EstimatedSalary']].values)





cbdf[[
 'CreditScore',
 'Age',
 'Tenure',
 'Balance',
 'NumOfProducts',
 'EstimatedSalary']]=X_standard
cbdf
#處理完後再分開
df_train=cbdf[0:trn]
df_train=pd.concat([df_train,df["Exited"]],axis=1)

df_test=cbdf[trn:]

df_train
df_test
```

```python
#分割訓練資料、測試資料
from sklearn.model_selection import train_test_split
all_in=df_train[['Gender_Female', 'Gender_Male', 'Geography_France', 'Geography_Germany',
       'Geography_Spain', 'RowNumber', 'CustomerId', 'CreditScore', 'Age',
       'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
       'EstimatedSalary']]
all_y=df_train['Exited']




(tr_i,te_i,tr_y,te_y)=train_test_split(all_in,all_y,train_size=0.8,random_state=1)
```

```python
#使用RandomForestClassifier 分割成5fold，遞迴
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier() 
rfecv = RFECV(estimator=clf_rf, step=1, cv=5,scoring='f1') 
rfecv = rfecv.fit(tr_i, tr_y)
```

```python
#使用RandomForestClassifier 分割成5fold，遞迴
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier() 
rfecv = RFECV(estimator=clf_rf, step=1, cv=5,scoring='f1') 
rfecv = rfecv.fit(tr_i, tr_y)

print('Optimal number of features :', rfecv.n_features_)
print('Best:', tr_i.columns[rfecv.support_])
```

```python
#將選出的特徵套用在模型
features=tr_i.columns[rfecv.support_]
tr_i=tr_i[features]
te_i=te_i[features]
clf_rf.fit(tr_i,tr_y)

p=clf_rf.predict(te_i)
from sklearn.metrics import classification_report 

print(classification_report(p, te_y))
```

```python
#計算分數
print(clf_rf.score(tr_i,tr_y))
print(clf_rf.score(te_i,te_y))
```
